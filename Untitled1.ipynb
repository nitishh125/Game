```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("unsdsn/world-happiness")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd

#List files in the download directory
print("Files in the dataset directory:")
for filename in os.listdir(path):
    print(filename)
```

```python
df = pd.read_csv(os.path.join(path, "2019.csv"))
print(df.head())
# Check shape (rows, columns)
print("Shape:", df.shape)

# Check data types
print("\nData Types:")
print(df.dtypes)

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Check duplicate rows
print("\nDuplicate Rows:", df.duplicated().sum())
```

1. The dataset contains df.shape[0] rows and df.shape[1] columns.
2. It includes different types of data such as integers, floats, and text (object) columns.
3. Missing values are identified using df.isnull().sum(), which shows the number of null values in each column.
4. Duplicate records are checked using df.duplicated().sum() to ensure data quality.
5. This dataset can be cleaned and analyzed further after handling missing values and duplicate entries.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(os.path.join(path, "2019.csv"))
# Handle missing values
# Fill numerical columns with median
num_cols = df.select_dtypes(include=["int64", "float64"]).columns
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Fill categorical columns with mode
cat_cols = df.select_dtypes(include=["object"]).columns
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

print(df.info())
print(df.head())

# Save cleaned dataset
df.to_csv("cleaned_dataset.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load dataset
df = pd.read_csv(os.path.join(path, "2019.csv"))

# Calculate counts for each country
country_counts = df["Country or region"].value_counts()

# Pie Chart
plt.figure(figsize=(10,10)) # Increased figure size for better (though still crowded) display
plt.pie(country_counts.values, labels=country_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribution of Countries")
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
# Load dataset
df = pd.read_csv(os.path.join(path, "2019.csv"))

# Heatmap
plt.figure(figsize=(8,6)) # Increased figure size for better readability
corr = df.select_dtypes(include=["int64", "float64"]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f") # Added format for annotations
plt.title("Correlation Heatmap of Numerical Features")
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
